```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Charger les données depuis un fichier CSV
df = pd.read_csv('data/cumulative-co-emissions.csv')
```

```python

df_filtered = df[df['Entity'].isin(['France', 'United Kingdom', 'United States', 'China', 'India' ,'Asia (excl. China and India)', 'Russia', 'European Union (27)'])]

total_emission_by_year = df_filtered.groupby('Year')['Annual CO₂ emissions'].sum()

df_filtered['Taux d\'émission (%)'] = df_filtered.apply(lambda x: (x['Annual CO₂ emissions'] / total_emission_by_year[x['Year']]) * 100, axis=1)
```

```python
df['Entity'].unique()
```

```python
plt.figure(figsize=(10, 6))

for country in df_filtered['Entity'].unique():
    subset = df_filtered[df_filtered['Entity'] == country]
    plt.plot(subset['Year'], subset['Taux d\'émission (%)'], label=country)

important_dates = [(1914, '1914'), (1929, '1929'), (1939, '1939'), (1973, '1973') ,(2008, '2008')]  # Exemple de dates clés
for date, label in important_dates:
    plt.axvline(x=date, color='gray', linestyle='--', linewidth=1)
    plt.text(date, plt.gca().get_ylim()[0], label, ha='center', va='top', rotation=90, color='gray', fontsize=8)


plt.xlabel('Année')
plt.ylabel('Taux d\'émission (%)')
plt.title('Taux d\'émission de CO2 des 9 principaux pays émetteurs en fonction de l\'émission mondiale')
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

Ce graphique représente la part d'émission de chaque pays en fonction de l'émission globale de CO2 dans le monde par année. On observe que quasiment 100 % des émissions de CO2 étaient produites par le Royaume-Uni entre les années 1750 et 1800. Cela s'explique par le fait que le Royaume-Uni est le premier pays à s'industrialiser, notamment grâce à l'invention de James Watt, la machine à vapeur qui augmenta la consommation de charbon dans le pays et lança la première révolution industrielle. On observe qu'à partir des années 1800, l'Europe commence elle aussi à émettre une partie du CO2 car elle suit le mouvement lancé par le Royaume-Uni. Les États-Unis, quant à eux, ne dépasseront l'Europe en terme d'émission de CO2 qu'à partir des années 1900. Nous avons choisi de relever certaines dates clés ayant eu un impact dans la production des pays. En effet, nous pouvons voir que chacune des deux guerres mondiales a entraîné une diminution du taux de CO2 en Europe mais augmentant celui des États-Unis. À la différence, on observe que la crise économique de 1929 a quant à elle provoqué une diminution du taux aux États-Unis. À partir des années 1950, on observe de nouveaux pays occupant une grande place dans l'émission de CO2, jusqu'à ce que la Chine devienne dans les années 2000 le plus gros émetteur de CO2. Cette augmentation auprès des pays d'Asie et cette diminution dans les pays occidentaux s'explique par un développement industriel et technologique des pays asiatiques et, à l'inverse, la diminution dans les pays occidentaux, notamment en Europe, s'explique par la délocalisation de nombreuses industries vers les pays asiatiques.

```python
import matplotlib.pyplot as plt
import pandas as pd

# Charger les données
df = pd.read_csv('co2.csv')

df3_filtered = df[df['Entity'].isin(['France', 'United Kingdom', 'Russia', 'United States', 'China', 'India' ,'Asia (excl. China and India)', 'European Union (27)'])]

# Tracer le graphique
plt.figure(figsize=(10, 6))

for entity in df3_filtered['Entity'].unique():
    subset = df3_filtered[df3_filtered['Entity'] == entity]
    plt.plot(subset['Year'], subset['Annual CO₂ emissions'], label=entity)

important_dates = [(1914, '1914'), (1929, '1929'), (1939, '1939'), (1973, '1973') ,(2008, '2008')]
for date, label in important_dates:
    plt.axvline(x=date, color='gray', linestyle='--', linewidth=1)
    plt.text(date, plt.gca().get_ylim()[0], label, ha='center', va='top', rotation=90, color='gray', fontsize=8)

plt.xlabel('Année')
plt.ylabel('Émissions annuelles de CO₂')
plt.title('Émissions annuelles de CO₂ en fonction de l\'année')
plt.legend()
plt.grid(True)
plt.show()
```

Ce graphique représente la quantité de CO2 émis par pays en fonction des années (en millions de tonnes). Cette courbes nous montres que la quantité d'émission de CO2 est en constante évolutions. En effet, on observe que le plus gros émetteur de CO2 dans les années 1900 émettait 12 fois moins que notre plus gros émetteur actuelle. Les seuls pays à avois baissés leurs émissions de C02 sont les pays occidentaux. Cependant, comme explique sur le graphe ci-dessus cette baisse s'explique par la désindustialisation de ces pays et la délocalisation de leurs usines. On oberve égalemment une augmentation très rapide dans certain comme la chine mais aussi l'inde depuis une vingtaines d'années.

```python
# Liste des noms à exclure
exclusions = [
    'Africa', 'Africa (GCP)', 'Asia', 'Asia (GCP)', 'Asia (excl. China and India)',
    'Central America (GCP)', 'Europe', 'Europe (GCP)', 'Europe (excl. EU-27)',
    'Europe (excl. EU-28)', 'European Union (27)', 'European Union (28)',
    'High-income countries', 'International aviation', 'International shipping',
    'Low-income countries', 'Lower-middle-income countries', 'North America',
    'Oceania', 'Oceania (excl. Australia and New Zealand)', 'South America',
    'Upper-middle-income countries', 'World', 'OECD (GCP)', 'Non-OECD (GCP)', 'North America (GCP)'
]

# Filtrer le DataFrame pour exclure les entrées de la liste des exclusions
df_only_country = df[~df['Entity'].isin(exclusions)]

df_only_continent = df[df['Entity'].isin(['Europe', 'Asia', 'North America', 'South America', 'Oceania', 'Africa'])]
df['Entity'].unique()

```

```python
# Calculer le total des émissions de CO2 pour chaque pays
total_emissions_by_continent = df_only_continent.groupby('Entity')['Annual CO₂ emissions'].sum()

# Trier les données pour avoir les pays avec les plus hautes émissions en premier
total_emissions_by_continent_sorted = total_emissions_by_continent.sort_values(ascending=False)

# Créer un graphique camembert
plt.figure(figsize=(10, 8))
plt.pie(total_emissions_by_continent_sorted, labels=total_emissions_by_continent_sorted.index, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Optionnel: Ajouter un titre
plt.title('Part des émissions de CO2 par continent')

plt.show()
```

```python
import plotly.express as px
import pandas as pd
import numpy as np

df2 = pd.read_csv('data/cumulative-co-emissions.csv')
df2 = df2.sort_values(by='cum_co2_em', ascending=True)

seuils = np.array([0, 3e6, 10e6, 30e6, 100e6, 300e6, 1e9, 3e9, 10e9,10e100])
labels = ['0 - 3M              ', '3M - 10M', '10M - 30M', '30M - 100M', '100M - 300M', '300M - 1B', '1B - 3B', '3B - 10B', '> 10B']
colors_scale = ['#FFFFCC', '#FFEDA0', '#FED976', '#FEB24C', '#FD8D3C', '#FC4E2A', '#E31A1C', '#BD0026', '#800026']
color_map = {label: color for label, color in zip(labels, colors_scale)}

df2['Category'] = pd.cut(df2['cum_co2_em'], bins=seuils, labels=labels, include_lowest=True)

df2['Color'] = df2['Category'].replace(color_map)

start_year = 1750
end_year = df2['Year'].max()
years = range(start_year, end_year + 1, 10) 

for year in years:
    data_1982 = df2[df2['Year'] == year]

    fig = px.choropleth(data_1982,
                        locations="Code",
                        color="Category",
                        color_discrete_map=color_map,
                        hover_data=["cum_co2_em"],
                        hover_name="Entity",
                        title=f"Emissions annuelles de CO2 en {year} par pays"
)
    fig.update_layout(geo=dict(showframe=False,
                            showcoastlines=True))

    fig.write_image(f'climate/assets/images/map/co2_{year}.png')
    #fig.show()
```

```python
import plotly.express as px
import pandas as pd
import numpy as np

data="data/annual-temperature-anomalies.csv"
df2 = pd.read_csv(data)

df2 = df2.sort_values(by='Temperature', ascending=True)

seuils = np.array([-555, -2, -1.5, -1, -0.5, 0, 0.5, 1, 1.5, 30])
labels = ['< -2','(-2) - (-1.5)','(-1.5) - 1','(-1) - (-0.5)', '(-0.5) - 0', '0 - 0.5', '0.5 - 1', '1 - 1.5', '> 1.5']
colors_scale = ['#4575b4', '#74add1', '#abd9e9', '#FEB24C', '#FD8D3C', '#FC4E2A','#E31A1C', '#BD0026', '#800026']
color_map = {label: color for label, color in zip(labels, colors_scale)}

df2['Température (°C)'] = pd.cut(df2['Temperature'], bins=seuils, labels=labels, include_lowest=True)

df2['Color'] = df2['Température (°C)'].replace(color_map)

start_year = 1750
end_year = df2['Year'].max()
years = range(start_year, end_year + 1, 10) 

for year in years:
    data_1982 = df2[df2['Year'] == year]

    fig = px.choropleth(data_1982,
                        locations="Code",
                        color="Température (°C)",
                        color_discrete_map=color_map,
                        hover_data=["Temperature"],
                        hover_name="Entity",
                        title=f"Anomalies des températures annuelles par pays en {year}",
                        category_orders={'Anomalie température (°C)': labels}
)
    fig.update_layout(geo=dict(showframe=False,
                            showcoastlines=True))

    fig.write_image(f'climate/assets/images/map/temperature_{year}.png')
```
