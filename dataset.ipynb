```python
import pandas as pd
```

```python
df = pd.read_csv('data/cumulative-co-emissions.csv')
df.head()
```

```python
global_tmp = pd.read_csv('data/GlobalLandTemperaturesByCountry.csv')
global_tmp.head()
```

```python
# Convertir la colonne 'dt' en datetime
global_tmp['dt'] = pd.to_datetime(global_tmp['dt'])

# Créer les colonnes 'year', 'month', 'day'
global_tmp['year'] = global_tmp['dt'].dt.year
global_tmp['month'] = global_tmp['dt'].dt.month
global_tmp['day'] = global_tmp['dt'].dt.day
global_tmp.head()
```

```python
# keep only 01 01 date
global_tmp = global_tmp[global_tmp['month'] == 1]
```

```python
global_tmp.head()
```

```python
df = df.rename(columns={'Entity': 'Country'})
df = df.rename(columns={'Year': 'year'})
df.head()
```

```python
# merge the two dataframes on the country column and the year column
merged = pd.merge(global_tmp, df, on=['Country', 'year'], how='left')
```

```python
africa = merged[merged['Country'] == 'Africa']
europe = merged[merged['Country'] == 'Europe']
asia = merged[merged['Country'] == 'Asia']
south_america = merged[merged['Country'] == 'South America']
north_america = merged[merged['Country'] == 'North America']
oceania = merged[merged['Country'] == 'Oceania']
```
