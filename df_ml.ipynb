```python
df = pd.read_csv('data/GlobalLandTemperaturesByCountry.csv')
df.head()
```

```python
# convert dt to year,month and day
df['dt'] = pd.to_datetime(df['dt'])
df['year'] = df['dt'].dt.year
df['month'] = df['dt'].dt.month
df['day'] = df['dt'].dt.day
```

```python
df.head()
```

```python
# Ok, I want to compress the data to have only one row per year and country
# I will calculate the average temperature for the year
df = df.groupby(['Country', 'year']).mean().reset_index()
df.head()
```

```python
df = df[df['year'] >= 1950]
```

```python
df[df['Country'] == 'Europe'].head(20)
```

```python
df2 = pd.read_csv('data/cumulative-co-emissions.csv')
df2.head()
```

```python
df1 = df.loc[df['Country'].isin(['Europe', 'Asia', 'Africa', 'North America', 'South America', 'Oceania'])]
df2 = df2.loc[df2['Entity'].isin(['Europe', 'Asia', 'Africa', 'North America', 'South America', 'Oceania'])]
```

```python
# merge df and df2 by Country and Year
newdf = pd.merge(df1, df2, left_on=['Country', 'year'], right_on=['Entity', 'Year'])
```

```python
newdf.drop(columns=['Code'], inplace=True)
newdf.head()
```

```python
# show unique country name
newdf['Country'].unique()
```

```python
newdf.isna().sum()
```

```python
newdf.to_csv('data/train.csv', index=False)
```
